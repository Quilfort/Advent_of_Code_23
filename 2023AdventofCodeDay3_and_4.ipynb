```python
# PART 1 any number adjacent to a symbol, even diagonally, is a "part number"
## Checken waarom for loop niet werkt
## Check if (start - 1 && end + 1) en  (i - 1 && 1 + 1) een symbol hebben

import re

with open('/content/drive/MyDrive/AdventOfCode2023/input3.txt') as f:
  matrix = []
  for i in f:
    matrix.append(i)

def checkIfValid(matrix, line_nb, start, end):

  pattern = re.compile(r'[^0-9.]')

  if start == 0:
    check_start = start
  else:
    check_start = start - 1

  if end == len(matrix[line_nb]) - 1:
    check_end = start
  else:
    check_end = end + 1

  checkString = matrix[line_nb][check_start:check_end]
  match = pattern.search(checkString)
  if match:
    return True

  if line_nb != 0:
    checkString = matrix[line_nb - 1][check_start:check_end]
    match = pattern.search(checkString)
    if match:
      return True

  if line_nb != len(matrix[line_nb]) - 1:
    checkString = matrix[line_nb + 1][check_start:check_end]
    match = pattern.search(checkString)
    if match:
      return True

  return False

total = 0

for line_nb, item in enumerate(matrix):
  print(matrix[line_nb])
  start = 0
  idx = 0
  end = 0
  i = 0
  addNumber = False

  while i < len(matrix[line_nb]) - 1:
    addNumber = False
    if matrix[line_nb][i].isdigit():
      start = i
      idx = start + 1
      while idx < len(matrix[line_nb]) and matrix[line_nb][idx].isdigit():
          idx += 1
      end = idx
      # print(matrix[line_nb][start:end])
      i = end
      addNumber = checkIfValid(matrix, line_nb, start, end)
      if addNumber:
        print(matrix[line_nb][start:end])
        total += int(matrix[line_nb][start:end])
    else:
      i += 1

  # if line_nb == 1:
  #   break

print("\nThis is Total:    ", total)
```

```python
## Part 2

import re

f = open('/content/drive/MyDrive/AdventOfCode2023/input4.txt')
f_content = f.read()
f.seek(0)

file_len = len(f.readlines())
cards = [0 for c in range(file_len + 1)]

f.seek(0)
total = 0
card_index = 0
# print("line_index:",  line_index)
for line in f.readlines():
    count = 0
    card_index = int(re.search("\d+", line.split(':')[0]).group(0))
    # print(line_index)
    trim_line = line[line.find(':') + 2:]
    trim_win = trim_line[:trim_line.find('|')]
    trim_card = trim_line[trim_line.find('|') + 2:]
    split_win = trim_win.split()
    split_card = trim_card.split()
    cards[card_index] += 1
    for win in split_win:
      for card in split_card:
        if win == card:
          count += 1

    for i in range(1, count + 1):
      cards[card_index + i] += cards[card_index]

total = sum(cards)
print(total)
```
