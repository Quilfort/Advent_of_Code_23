```python
# Part 1 refactor

import re

file_data = open('/content/drive/MyDrive/AdventOfCode2023/input5.txt').read().splitlines()
# print(file_data)
maps = []
for line in file_data[2:]:
  if 'map' in line:
    maps.append(dict())
  elif line != '':
    dest, src, len = [int(value) for value in line.split()]
    maps[-1][range(src, src+len)] = range(dest, dest+len)

def find_location(init_value: int) -> int:
  value = init_value
  for current_map in maps:
    value = next(
            (dest_range.start + (value - src_range.start)
             for src_range, dest_range in current_map.items()
             if value in src_range),
            value
        )
    return value

seeds = [int(seed) for seed in re.findall(r'\d+', file_data[0])]
locations = [find_location(seed) for seed in seeds]

print(min(locations))
```

```python
## Part 1
file_data = open('/content/drive/MyDrive/AdventOfCode2023/input6.txt').read().splitlines()
times = [int(value) for value in file_data[0].split()[1:]]
dists = [int(value) for value in file_data[1].split()[1:]]
answer = 0
count = 0

def beat_distance(time, distance):
  ret_count = 0
  for i in range(time):
    check = (time - i) * i
    if (check > distance):
      ret_count += 1
  return ret_count

for index, dist in enumerate(dists):
  count = 0
  count = beat_distance(times[index], dist)
  if index == 0:
    answer = count
  else:
    answer *= count

print(answer)
```

```python
## Part 2

def beat_distance(time, distance):
  ret_count = 0
  for i in range(time):
    check = (time - i) * i
    if (check > distance):
      ret_count += 1
  return ret_count

file_data = open('/content/drive/MyDrive/AdventOfCode2023/input6.txt').read().splitlines()
times = [int(value) for value in file_data[0].split()[1:]]
dists = [int(value) for value in file_data[1].split()[1:]]
times_combined = int(''.join(map(str, times)))
dists_combined = int(''.join(map(str, dists)))

print(beat_distance(times_combined, dists_combined))
```
