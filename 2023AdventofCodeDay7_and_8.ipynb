```python
##Part 1
file_data = open('/content/drive/MyDrive/AdventOfCode2023/input7.txt').read().splitlines()
string_str = "AKQJT98765432"
win_list = []
end_list = []
answer = 0

### Check the wins

def three_two(hand):

  for i in range(2):
    count = hand.count(hand[i])
    if count == 3:
      return 4
  return 3

def four_full(hand):

  count_first = hand.count(hand[0])
  if count_first == 1 or count_first == 4:
    return 6 #Four of a Kind
  else:
    return 5 #full house

def check_type(hand):
  store_hand = set(hand)
  unique = len(store_hand)

  if unique == 1:
    return 7 #Five of a kind
  if unique == 2:
    return (four_full(hand))
  if unique == 3:
    return (three_two(hand))
  if unique == 4:
    return 2 #One Pair
  return 1 #highest card

## Sorted List check

def sort_strongest(num_list):
  strongest = sorted(num_list, key=lambda x: [string_str.index(char) for char in x[0][:5]], reverse=True)
  return strongest

## MAIN
for index, game in enumerate(file_data):
  player = game.split()
  win = check_type(player[0])
  new_line = file_data[index] + " " + str(win)
  split_line = new_line.split()
  win_list.append(split_line)

sorted_win_list = sorted(win_list, key=lambda x: int(x[2]), reverse=True)

for i in range(1, 8):
  str_num = str(i)
  get_num_win_list = [num_list for num_list in sorted_win_list if str(num_list[2]).endswith(str_num)]
  sorted_num_list = sort_strongest(get_num_win_list)
  for x in sorted_num_list:
    end_list.append(x)

print(end_list)
print("--------")

for index, end in enumerate(end_list):

  game = end_list[index]
  bid = int((end_list[index][1]) )
  mult = (index + 1)
  win = bid * mult
  print(game, bid, mult, win)
  answer += win

print(answer)



```
